```python
images_dir = "/content/images/images/"
images_cleaned = "/content/images/cleaned"
```

```python
import cv2
import pytesseract
from PIL import Image, ImageDraw
import os

def remove_text_from_image(image_path):

    image = cv2.imread(image_path)

    h, w, _ = image.shape
    boxes = pytesseract.image_to_boxes(image)

    for box in boxes.splitlines():
        box = box.split(' ')
        x, y, x2, y2 = int(box[1]), int(box[2]), int(box[3]), int(box[4])
        y = h - y
        y2 = h - y2

        image = cv2.rectangle(image, (x, y), (x2, y2), (255, 255, 255), -1)

    output_path = os.path.join(images_cleaned, 'cleaned_' + os.path.basename(image_path))
    cv2.imwrite(output_path, image)
    print(f"Текст удален с изображения {image_path}, результат сохранен как {output_path}")

for filename in os.listdir(images_dir):
    if filename.endswith(".png") or filename.endswith(".jpg"):
        image_path = os.path.join(images_dir, filename)
        remove_text_from_image(image_path)
```
